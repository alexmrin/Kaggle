```python
import pandas as pd
```

```python
practice = pd.read_csv('/content/drive/MyDrive/pandas practice/pandas_demo.csv')
```

```python
practice
```

```python
practice.city = practice.city.map(lambda cty: cty.upper())
```

```python
def experience(row):
  if row.age < 25:
    return 'Junior'
  elif row.age < 30:
    return 'Mid-level'
  else:
    return 'Senior'

experience_level = practice.apply(experience, axis=1)
practice['experience_level'] = experience_level
practice
```

```python
avg = practice.salary.mean()
f'average salary is ${avg:.2f}'
oldestID = practice.age.idxmax()
f'oldest person is {practice.loc[oldestID, "name"]} and is {practice.loc[oldestID, "age"]} years old.'
youngestID = practice.age.idxmin()
f'youngest person is {practice.loc[youngestID, "name"]} and is {practice.loc[youngestID, "age"]} years old.'
```

```python
avg_salaries = practice.groupby(['experience_level']).salary.mean()
avg_salaries
```

```python
'''Add a new column named job_city which combines the job and city columns. The format should be "Job in City" (e.g., "Engineer in New York").'''
```

```python
practice['job_city'] = practice.job + " in " + practice.city.map(lambda x: x.title())
practice
```

```python
'''How many people have a salary above the overall average and are classified as 'Mid-level'?'''
```

```python
overall_avg = practice.salary.mean()

def above_avg(person):
  if person.salary > overall_avg:
    return 1
  else:
    return 0

mid_level = practice.loc[(practice.experience_level == 'Mid-level')]
top_mid_level = mid_level.apply(above_avg, axis=1)
top_mid_level = top_mid_level.sum()
top_mid_level
```

```python
'''Find the city with the highest number of employees. Calculate the median age of employees in each city.'''
```

```python
highest_city = practice.city.value_counts()
max_num = highest_city.max()
highest_cities = highest_city.loc[highest_city == max_num].index.map(lambda x: x.title()).tolist()
print(highest_cities)

median_ages = practice.groupby('city').age.median()
print(median_ages)
```

```python
'''Subtract the mean salary from each individual's salary and store the result in a new column named salary_deviation_from_mean. '''
```

```python
# overall_avg
practice['salary_deviation_from_mean'] = practice.salary - overall_avg
practice
```

```python
'''Find out who (name) has the highest salary among the 'Junior' experience level.'''
```

```python
highest_junior = practice.loc[practice.experience_level == 'Junior']['salary'].idxmax()
highest_junior = practice.loc[highest_junior, 'name']
highest_junior
```

```python
'''For each experience level, find the city that has the highest average salary.'''
```

```python
mean_salaries = practice.groupby(["experience_level", "city"]).salary.mean().reset_index()
ranked_salaries = mean_salaries.sort_values(by=['experience_level', 'salary'], ascending=[True, False])
top_cities = ranked_salaries.drop_duplicates(subset=['experience_level'])
top_cities = top_cities.city.map(lambda x: x.title()).tolist()
top_cities
```

```python
'''Create a new column named salary_category: If the salary is in the top 25% of all salaries, label as 'High'. If the salary is in the bottom 25% of all salaries, label as 'Low'. Otherwise, label as 'Medium'.'''
```

```python
top = practice.salary.quantile(.75)
bottom = practice.salary.quantile(.25)
def salary_categorize(person):
  if person.salary >= top:
    return 'High'
  elif person.salary <= bottom:
    return 'Low'
  else:
    return 'Medium'

practice['salary_category'] = practice.apply(salary_categorize, axis=1)
practice
```
